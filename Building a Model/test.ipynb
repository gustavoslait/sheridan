### Orange Inc. wants to maximize the profit for their two Juice products X and Y which are sold at $8 and $12 respectively. Orange Inc has 1200 resource units available every day in their factory; product X requires 10 units while Y requires 5 units. Both products require a production time of 5 minutes and total available working hours are 8 in a day.


```r
install.packages("lpSolve")
library(lpSolve)
```

### Estoy importando las libreria lpsolve que es una libreria de R disenada para resolver problemas de programcion lineal y prog entera. Con esta libreria podremos encontrar soluciones optimas de maximo o minimo de una funcion objetiva sujeta a restricciones lineales.

```r
# Graficar problemas LP 2 variables: región factible + óptimo
if (!requireNamespace("ggplot2", quietly = TRUE)) {
  install.packages("ggplot2", repos="https://cloud.r-project.org")
}

library(ggplot2)

# Parámetros ya definidos
obj <- c(8, 12)
A <- matrix(c(10,5,5,5), nrow = 2, byrow = TRUE)
b <- c(1200, 480)

grid <- expand.grid(x1 = seq(0, 150, length.out = 300),
                    x2 = seq(0, 150, length.out = 300))

grid$cons1 <- A[1,1]*grid$x1 + A[1,2]*grid$x2
grid$cons2 <- A[2,1]*grid$x1 + A[2,2]*grid$x2
grid$feasible <- (grid$cons1 <= b[1]) & (grid$cons2 <= b[2])

grid$z <- obj[1]*grid$x1 + obj[2]*grid$x2

res <- lp("max", obj, A, c("<=","<="), b)
opt <- data.frame(x1 = res$solution[1], x2 = res$solution[2], z = res$objval)

p <- ggplot(grid, aes(x = x1, y = x2)) +
  geom_raster(aes(fill = feasible), interpolate = TRUE, alpha = 0.4) +
  scale_fill_manual(values = c("FALSE" = "white", "TRUE" = "lightblue"), guide = FALSE) +
  geom_contour(aes(z = z), breaks = seq(0, max(grid$z), by = 50), color = "gray50", alpha = 0.5) +
  geom_point(data = opt, aes(x = x1, y = x2), color = "red", size = 4) +
  geom_text(data = opt, aes(x = x1, y = x2, label = sprintf("opt (%.1f, %.1f) z=%.1f", x1, x2, z)),
            vjust = -1, color = "red") +
  labs(title = "Región factible y solución óptima LP",
       subtitle = "max 8x1 + 12x2 sujeto a 10x1+5x2<=1200, 5x1+5x2<=480",
       x = "x1", y = "x2") +
  coord_cartesian(xlim = c(0, 150), ylim = c(0, 150)) +
  theme_minimal()

print(p)
```
