```r
# Ver versión R y cargar IRkernel
R.version.string

# Instala lpSolve si aún no está instalado
if (!requireNamespace("lpSolve", quietly = TRUE)) {
  install.packages("lpSolve", repos="https://cloud.r-project.org", dependencies=TRUE)
}
library(lpSolve)

# Comprobar lpSolve
print(packageVersion("lpSolve"))

# Ejemplo simple de programación lineal
f.obj <- c(2, 3, 4)

f.con <- matrix(c(
  3, 2, 1,
  2, 5, 3
), nrow = 2, byrow = TRUE)

f.dir <- c("<=", "<=")
f.rhs <- c(10, 15)

res <- lp("max", f.obj, f.con, f.dir, f.rhs)

# Resultado
print(res$status)      # 0=óptimo
print(res$objval)      # valor objetivo
print(res$solution)    # x1, x2, x3

# Opcional: pasa resultados a df
df_out <- data.frame(
  variable = paste0("x", 1:length(res$solution)),
  value = res$solution
)
print(df_out)

# Guardar output en CSV
write.csv(df_out, "LP_solution.csv", row.names = FALSE)
```

```r
objection.in<-c(8,12) 
const.mat<-matrix(c(10,5,5,5), nrow=2, byrow=TRUE)                 
time_constrait<-(8*60)  
resource_constrait<-1200
const.rhs<-c(resource_constrait,time_constrait)
const.dir<-c("<=","<=")
```

```r
optimum_profit <- lp(
  direction = "max",
  objective.in = objection.in,
  const.mat = const.mat,
  const.dir = const.dir,
  const.rhs = const.rhs
)

optimum_profit$status    # 0 debería ser óptimo
optimum_profit$objval
optimum_profit$solution
```
